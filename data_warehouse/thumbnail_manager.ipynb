```python
import pandas as pd
import requests
import os
```

```python
product_data_id = 'MLM1000_1000'
df = pd.read_csv(f'./products/{product_data_id}.csv')
```

```python
for index, row in df.iterrows():
    # Get the catalog_product_id and thumbnail URL from the row
    catalog_product_id = row['catalog_product_id']
    thumbnail_url = row['thumbnail']
    
    # Download the thumbnail image
    response = requests.get(thumbnail_url)
    
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Save the image with the desired filename in the 'thumbnails' folder
        filename = f'thumbnails/{catalog_product_id}_thumbnail.jpg'  # Assuming thumbnails are JPEG images

        # Check if the file already exists
        if os.path.exists(filename):
            print(f"Thumbnail {catalog_product_id} already exists. Skipping...")
            continue

        with open(filename, 'wb') as f:
            f.write(response.content)
        print(f"Thumbnail {catalog_product_id} downloaded successfully.")
    else:
        print(f"Failed to download thumbnail {catalog_product_id}. Status code:", response.status_code)
```
